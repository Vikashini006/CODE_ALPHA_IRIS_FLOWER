```python
from sklearn.model_selection import train_test_split
from sklearn import metrics
from sklearn.tree import DecisionTreeClassifier
```

```python
feature_cols=df.select_dtypes(include=["float"]).columns
```

```python
x=df[feature_cols]
y=df["target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.3,random_state=1)
```

```python
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
clf=DecisionTreeClassifier()
```

```python
clf.fit(x_train,y_train)
y.pred=clf.predict(x_test)
```

```python
print("Accuracy:",metrics.accuracy_score(y_test,y.pred))
```

```python
from sklearn.metrics import classification_report
report=classification_report(y_test,y.pred)
print("Classification report:\n",report)
```

```python
from sklearn import tree
plt.figure(figsize=(12,8))
tree.plot_tree(clf.fit(x_train,y_train))
```

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, classification_report

S = 5  
knn = KNeighborsClassifier(n_neighbors=S) 
knn.fit (x_train, y_train) 
y_pred = knn.predict(x_test) 

accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)

report_KNN = classification_report(y_test, y_pred)  
print("Classification Report:\n", report_KNN)
```
